## 1- Tracage de Base:
On peut tracer des fonctions mathematiques simples a une variable en utilisant la methode "$plot$", i.e. En utilisant $plot(x,y)$, on va generer une figure dans laquelle on trace la fonction "y" en fonction de "x". En effet les arguments de '$plot$' sont des "arrays" ou des "listes" de meme taille. On peut encore utiliser $plot(y)$ pour tracer y en fonction de l'indice du array ou de la liste en question, i.e. $plot(y)$ est equivalente a $plot(range(len(y)), y)$
- Exemples:

```python
# Tracer la fonction y(x) = exp(-x**2)*cos(x) dans un intervalle [-1, 1]
import numpy as np
import matplotlib.pyplot as plt

def y(x):
    return np.exp(-x**2)*np.cos(x)

x = np.linspace(-np.pi, np.pi, 1000) #  Generer un array [-pi, pi] avec 1000 points et donc un espacement de 0.001

plt.plot(x,y(x))
plt.title("Un premier plot")
plt.grid()
#plt.show() # Optionnel en IPython mais obligatoire dans un script Python ordinaire. 
```

```python
# Un trace avec des points en chaque 100e point de 'x':
points = x[::100] # La serie de points
plt.plot(x,y(x))
plt.plot(points,y(points), 'r*') # Markeurs rouges en forme d'etoile
plt.title("Un second plot")
plt.grid()
```

Comme on peut le voir, la fonction plot peut addmettre plusieurs arguments, la couleur standard d'un plot est le bleu ciel. Les axes sont mis a l'echelle automatiquement, le style du plot peut etre exactement specifiee apres les deux premieres entrees ! La commande "title" donne un titre au plot. 

L'utilisation multiple de plot superpose les plots sur une meme figure, si on ne veut pas les vois dans une seule et meme figure, il suffit de definir plus qu'une seule figure:

```python
# Reprenons le meme programme precedent avec une autre fonction g(x):
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x**2)*np.cos(x)

def g(x):
    return np.exp(-x**2)*np.sin(x)

x = np.linspace(-np.pi, np.pi, 1000) #  Generer un array [-pi, pi] avec 1000 points et donc un espacement de 0.001

plt.plot(x,f(x))
plt.plot(x,g(x))
plt.title("Un troisieme plot")
plt.grid()
```

On remarque premierement que les couleurs sont assignes automatiquement. Pour separer les traces, on peut utiliser la commande figure():

Les traces multiples sur une meme figure peuvent etre expliquees en utilsant la commande 'legend':
- Exemple: Ajustement par des polynomes(Fitting): Pour cela on utilise les commandes polyfit et polyval qui appartiennent a la bibliotheque NumPy,

```python
xi = range(5)
y = np.array([1, 2, 1, 3, 5])
p2 = np.polyfit(xi,y,2) # Fit de y(x) avec un polynome d'ordre 2
p4 = np.polyfit(xi,y,4) # Fit de y(x) avec un polynome d'ordre 4
# Tracer les polynomes qui 'fittent' le mieux les donnes y(xi) avec des "labels" expliquant chaque courbe.
x = np.linspace(-1, 5, 200)
plt.plot(x, np.polyval(p2, x), label= "Fit avec un polynome d'ordre 2")
plt.plot(x, np.polyval(p4, x), label= "Fit avec un polynome d'ordre 4")

plt.plot(xi,y, "r*")
plt.grid()
plt.show()
```

Comme on voit dans l'exemple precendent les deux courbes ne sont pas tres claires ainsi que les points y(xi), et les labels des courbes ne sont pas representes. On va remedier a ces deux problemes:

```python
xi = range(5)
y = np.array([1, 2, 1, 3, 5])
p2 = np.polyfit(xi,y,2) # Fit de y(x) avec un polynome d'ordre 2
p4 = np.polyfit(xi,y,4) # Fit de y(x) avec un polynome d'ordre 4
# Tracer les polynomes qui 'fittent' le mieux les donnes y(xi) avec des "labels" expliquant chaque courbe.
x = np.linspace(-1, 5, 200)
plt.plot(x, np.polyval(p2, x), label= "Fit avec un polynome d'ordre 2")
plt.plot(x, np.polyval(p4, x), label= "Fit avec un polynome d'ordre 4")

plt.plot(xi,y, "*")

## Resolution:
# Pour afficher les labels, on utilise la commande 'legend':
plt.legend(loc = "upper left") # loc = localisation de la legende, 
# Pour ajuter les axes:
plt.axis([-1, 5, 0, 6]) # [x0, x1, y0, y1]
##############

plt.grid()
plt.show()
```

### 2.1- Les histogrames:
On peut, au lieu de tracer des simples plots, tracer des histogrames ! Pour cela, on utilise la fonction 'hist':

```python
import numpy as np
import matplotlib.pyplot as plt

# random vector with normal distribution
sigma, mu = 2, 10
x = sigma*np.random.standard_normal(10000)+mu
plt.hist(x,50,normed=1)
z = np.linspace(0,20,200)
plt.plot(z, (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(z-mu)**2/(2*sigma**2)),'g')

# title with LaTeX formatting
plt.grid()
plt.title('Histogram with $\mu = ${:} and $\sigma = ${:}'.format(mu,sigma))
plt.show()
```

### 2.2- Les subplots:

On pourrait avoir besoin de tracer plus qu'un seul plot dans une et meme figure, pour cela, on devra utiliser des sous-plots, des "subplots" en language PYTHON.
- Exemple: Onde progressive a plusieurs instants "t".

```python
import numpy as np
import matplotlib.pyplot as plt

def wave(x, t):
    (c, x0, A) = (1.0, 1.0, 1.0)
    X = (x - c*t)/x0
    return A*np.exp(-(X**2)/2)

x = np.linspace(-2*np.pi, 2*np.pi, 1000)


plt.subplot(4, 1, 1)  # (4, 1) 4 lignes et une colonne de subplots, puis on choisi le subplot 1 pour tracer notre figure.
plt.plot(x, wave(x, 0), 'r') # rouge et solid
plt.grid(True)  # Une grille pour chaque subplot

plt.subplot(4, 1, 2)  # (4, 1) 4 lignes et une colonne de subplots, puis on choisi le subplot 2 pour tracer notre figure.
plt.plot(x, wave(x, 1.0), 'b--') # bleu et dashed
plt.grid(True)  # Une grille pour chaque subplot

plt.subplot(4, 1, 3)  # (4, 1) 4 lignes et une colonne de subplots, puis on choisi le subplot 3 pour tracer notre figure.
plt.plot(x, wave(x, 2.0), 'g:') # vert et dotted
plt.grid(True)  # Une grille pour chaque subplot

plt.subplot(4, 1, 4)  # (4, 1) 4 lignes et une colonne de subplots, puis on choisi le subplot 4 pour tracer notre figure.
plt.plot(x, wave(x, 3.0), 'k-.') # Noir et dashed-dotted
plt.grid(True)  # Une grille pour chaque subplot

# Sauvegarder une figure:
plt.savefig("/home/mouloud/Desktop/Doctorat/Modelisation/Cours_Complet_Jupyter/Figures/Save.pdf", dpi = 900) 
# Sauvegarde d'une figure sous forme PDF avec 900 pts par pouce

plt.show()
```

- Remarques:
    - On voit que l'axe des x n'apparait pas dans les 3 premiers subplots. 
    - L'axe des x est commun pour tout les subplots et donc on peut l'adopter une seule fois ! Ceci va etre effectuee apres introduction des objets "axes" dans une section avancee de ce chapitre.

```python
# Pour remedier au premier probleme, on utilise: subplots_adjust(hspace = 0.7) par exemple !
import numpy as np
import matplotlib.pyplot as plt

def wave(x, t):
    (c, x0, A) = (1.0, 1.0, 1.0)
    X = (x - c*t)/x0
    return A*np.exp(-(X**2)/2)

x = np.linspace(-2*np.pi, 2*np.pi, 1000)


plt.subplot(4, 1, 1)  # (4, 1) 4 lignes et une colonne de subplots, puis on choisi le subplot 1 pour tracer notre figure.
plt.plot(x, wave(x, 0), 'r') # rouge et solid
plt.grid(True)  # Une grille pour chaque subplot

plt.subplot(4, 1, 2)  # (4, 1) 4 lignes et une colonne de subplots, puis on choisi le subplot 2 pour tracer notre figure.
plt.plot(x, wave(x, 1.0), 'b--') # bleu et dashed
plt.grid(True)  # Une grille pour chaque subplot

plt.subplot(4, 1, 3)  # (4, 1) 4 lignes et une colonne de subplots, puis on choisi le subplot 3 pour tracer notre figure.
plt.plot(x, wave(x, 2.0), 'g:') # vert et dotted
plt.grid(True)  # Une grille pour chaque subplot

plt.subplot(4, 1, 4)  # (4, 1) 4 lignes et une colonne de subplots, puis on choisi le subplot 4 pour tracer notre figure.
plt.plot(x, wave(x, 3.0), 'k-.') # Noir et dashed-dotted
plt.grid(True)  # Une grille pour chaque subplot

# Sauvegarder une figure:
#plt.savefig("/home/mouloud/Desktop/Doctorat/Modelisation/Cours_Complet_Jupyter/Figures/Save.pdf", dpi = 900) 
# Sauvegarde d'une figure sous forme PDF avec 900 pts par pouce

plt.subplots_adjust(hspace = 0.7)
plt.show()
```

### 2.1- Les meshgrids (Grilles) et les contours:
- On voudrait representer graphiquement une fonction scalaire sur un rectangle:

\begin{equation}
f: [a, b] \times [c, d] \rightarrow \mathbf{R}
\end{equation}

- Pour ce faire, on doit generer une grille rectangulaire sur $[a, b] \times [c, d]$. Ceci est effectuee par la commande "meshgrid":

```python
x = np.linspace(0.0, 3.0, 1000) # 1000 est le nombre de points de discretization selon l'axe des 'x'
y = np.linspace(0.0, 1.0, 500)  # 500 est le nombre de points de discretization selon l'axe des 'y'

X, Y = np.meshgrid(x, y)
```

#### Remarques:
- X et Y sont des 'arrays' de la forme (n, m) de telle sorte que (X[i, j], Y[i, j]) contiennent les coordonnees du point $P_{ij}$ de la grille comme le montre la figure suivante:

- Un rectangle discretizee par "meshgrid" va etre utilisee pour visualiser le comportement d'une iteration. Mais, dans un premier lieu, on va l'utiliser pour tracer les courbes de niveau d'une fonction. Ceci est effectuee par la commande 'contour'.
- Exemple: La fonction de Rosenbrock:
    - La fonction est definie par:
    \begin{align*}
    f: [-0.5&, 2.0] \times [-1.5, 4.0] \rightarrow \mathbf{R} \\
    f(x,y) &= (1-x)^{2} + 100\:(y - x^{2})^{2} 
    \end{align*}
    - Courbes de niveau:

```python
Fonc = lambda x,y: 2*(np.exp(-X**2 - Y**2) - np.exp(-(X - 1)**2 - (Y - 1)**2))

x = np.linspace(-3.0, 3.0, 200)
y = np.linspace(-2.0, 2.0, 200)

X, Y = np.meshgrid(x, y)
Z = Fonc(X, Y)
plt.contour(X, Y, Z,cmap = 'gray')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Remarques:
- Afin de garder le code compact, une fonction anonyme indiquee par le mot-cle "lambda" est utilisee.
- La fonction anonyme utilisant le mot-cle lambda est expliquee dans le chapitre sur les fonctions dans la section 'Fonctions anonymes'
- Si les niveaux a tracer ne sont pas specifiees comme arguments a la fonction 'contour', alors la fonction va automatiquement specifier les niveaux.
- La fonction 'contourf' effectue la meme tache que fait 'contour' mais remplit le plot par des couleurs qui indiquent les differents niveaux.

```python
# Meme fonction avec 'contourf'
Fonc = lambda x,y: 2*(np.exp(-X**2 - Y**2) - np.exp(-(X - 1)**2 - (Y - 1)**2))

x = np.linspace(-3.0, 3.0, 200)
y = np.linspace(-2.0, 2.0, 200)

X, Y = np.meshgrid(x, y)
Z = Fonc(X, Y)
plt.contourf(X, Y, Z)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
