On voudrait afficher le rendu sous forme exponentielle avec une certaine precision, pour cela on peut utiliser des arguments $X$ tq $\{:\}\: \longrightarrow \:\{:X\}$
- Exemple:

```python
a = 33.4568
print("1/ a = {:}\n".format(a))     # tel qu'il est

print("2/ a = {:f}\n".format(a))    # Ces deux sont
print("3/ a = {:1f}\n".format(a))   # identiques

print("4/ a = {:1.1f}\n".format(a)) # Precision de une valeurs apres la virgule
print("5/ a = {:1.3f}\n".format(a)) # Precision de trois valeurs apres la virgule

print("6/ a = {:e}\n".format(a))     # Affichage exponentiel
print("7/ a = {:1.3}\n".format(a))  # Affichage exponentiel avec une precision de deux valeurs apres la virgule avant 
                                   # l'expo
```

- Un Exemple expliquant lui meme:

```python
Prix = 30.512147
print("Le {Nom} est {Valeur:1.5f}".format(Nom = "Prix", Valeur = Prix))
```

### 2.4.1- Parties reelles et imaginaires d'un nombre complexe:
- Exemple:

```python
z = 1 + 2j
print("Partie reelle: ", z.real)
print("Partie imaginaire: ", z.imag)
```

Ce qui est tres interessant, c'est que les methodes "$\textit{real}$" and "$\textit{imag}$" fonctionnent bien pour les nombres complexes ordinaires que pour les tableaux complexes de $\textrm{NumPy}$.
On peut faire une petite demonstration de cette propriete a travers un exemple:
Dans cet exemple, on va calculer les "$N$" racines de l'unite qui sont donnes par: $\:z_k = \exp\left({\frac{2j\pi k}{N}}\right)$ pour $k = 0, ..., N-1$. Il ne serai pas necessaire de comprendre maintenant le programme qu'on va presenter !

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 1000)

def cercle_unite(x, x0, y0):
    X = x - x0
    y1 = y0 + np.sqrt(1 - X**2)
    y2 = y0 - np.sqrt(1 - X**2)
    
    return y1, y2


def racines(N, k):
        xr = np.cos(2*np.pi*k/N)
        yr = np.sin(2*np.pi*k/N)
        return xr, yr
    
y1, y2 = cercle_unite(x, 0, 0)

fig, ax = plt.subplots()
(x0, y0) = (0, 0)
y1, y2 = cercle_unite(x, x0, y0)

ax.plot(x, y1, color = "blue", linestyle = "dotted")
ax.plot(x, y2, color = "blue", linestyle = "dotted")

N = 10

for k in range(N):
    xr, yr = racines(N, k)
    ax.scatter(xr, yr, color = "red", marker = "o", edgecolors = "black")
    ax.set_xlabel("Axe des reels")
    ax.set_ylabel("Axe des imaginaires")

plt.title("Les $N$ racines de l'unite")
plt.grid(True, linestyle = "dotted")
plt.show()
```
