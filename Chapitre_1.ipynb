- ### Le "while" de PYTHON: On repete une portion de code jusqu'a ce que la condition donnee soit verifiee.
    - Syntaxes:
        * while condition:
                Bloc_d'instructions
                
    - Exemple: Programme qui cree une liste d'entiers qui commence par 5 et se termine par 15.
    
    Algorithme: 
    
        1- On cree une variable et on lui affecte la valeur 5. Et on cree une liste vide.
        
        2- On la stocke dans un tableau puis on incremente par 1 la derniere valeur

```python
n = 5
liste = []
while n < 16:
    liste.append(n) # Apport de n a la liste
    n = n + 1 # Incrementation
print(liste)    
```

## 9- Traitement des erreurs et exceptions:
La gestion des erreurs d'execution des parties sensibles d'un code est importante, cette partie presente comment on gere les erreurs et les exceptions qu'on peut parcourir sur PYTHON.

Lorsqu'une erreur se produit, elle traverse toutes les couches de code comme une bulle d'air qui remonte a la surface de l'eau. Si elle atteint la surface sans etre interceptee par le mecanisme des 'exceptions', le programme s'arrete et l'erreur est affichee, le 'traceback', qui est un message d'erreur complet qui s'affiche precisant l'ensemble des couches traversees.

---> Gerer une exception = Intercepter une erreur possible pour eviter l'arret d'un programme

Une exception separe d'un cote la sequence d'instructions a executer quand tout se passe bien, et d'un autre cote, une ou plusieurs instructions a executer en cas d'erreur.

    - Syntaxe: La sequence normale d'instructions est placee dans un bloc "try"
               Si une erreur est detectee (levee d'exception), elle est traitee dans le bloc "except" approprie 
               (Le gestionnaire d'exception)
               
  - Exemple: le sinus cardinal est defini par: $\:sinc(x) = \frac{sin(x)}{x}$, pour $x = 0$, il y'a un probleme de divergence mais on sait que $sinc(0) = 1$, Que faire pour eviter l'erreur ? 

```python
from math import sin

for x in range(-4, 5):
    try:
        print("{:.3f}".format(sin(x)/x), end = " ")  # .3f -->  3 nombres apres la virgule.
        # end = " ", pour ne pas sauter de ligne
    except ZeroDivisionError:  # On fournit toujours un type d'exception
        print("1.0", end = " ")
```

Toutes les exceptions levees par PYTHON appartiennent a la meme famille nommee $\textit{Exception}$
##### - Exemples d'exceptions PYTHON:    
- ZeroDivisionError
- NameError
- TypeError
- ValueError
- RuntimeError etc....

On peut aussi definir nos propres types d'exceptions comme on l'avait fait dans l'exemple precedent. La syntaxe d'une exception est comme suit:

```python
# Portion de code non executable !
try:
    "......" # Sequence normale d'execution
except <exception_1> as e1:
    "......" # Traitement de l'exception 1
except <exception_2> as e2:
    "......" # Traitement de l'exception 2    
"""
etc...
"""
else: 
    # Clause executee en cas d'absence d'erreur
finally:
    # Clause toujours executee en tout cas
```

- Exemple:

```python
def divide(x, y):
    try:
        resultat = x / y
    except ZeroDivisionError:
        print("- Division par Zero impossible")
    else:
        print("- Le resultat est: {:}".format(resultat))
    finally:
        print("Programme termine\n") # \n est pour le saut de ligne

divide(8, 2)
divide(8, 0)
```

##        10- Fonctions, Modules et Espace-noms en bref:
### a- Les fonctions:

Une fonction est un ensemble d'instructions regroupees sous un nom et s'executant a la demande. Elle englobe un ensemble de lignes de codes qui sont susceptibles a se repeter a plusieurs reprises dans le programme. Une fonction est donc une sorte de "factorisation de code".

- Syntaxe:

```python
# Code non executable
def Nom_de_la_fonction(arg1, arg2, arg3, ..., argn):
    """
    Documentation facultative (mais fortement recomendee) de la fonction
    """
    Ensemble de 
    lignes de code
    a executer
```

- Exemple de fonction: Une fonction qui calcul a la demande la somme, la difference, le produit ou le quotient de deux nombres a la demande:

```python
entree = input("Choisissez une des operations suivantes:\n\t1- +\n\t2- -\n\t3- *\n\t4- /\nVotre choix: ")
if (entree == "+" or entree == "-" or entree == "*" or entree == "/"):
    pass # Depasser cette ligne si la condition est vraie
else:
    raise ValueError("Choix erronne")
    
def SP(a, b):
    if entree == "+":
        return a + b
    elif entree == "-":
        return a - b
    elif entree == "*":
        return a*b
    else: 
        try:
            return a/b
        except ZeroDivisionError:
            raise ValueError("b ne peut pas etre nul, Resultat -> Infini")
a = float(input("Entrez la valeur de a: "))
b = float(input("Entrez la valeur de b: "))
Resultat = SP(a, b)
print("Le resultat est: {:} {:} {:} = {:}".format(a, entree, b, Resultat))
```

### a.1. Passage des arguments:
 - Passage par affectation: Chaqu'un des parametres de la fonction correspond dans l'ordre a un argument de celle-ci comme dans l'exemple precedent.
 - Passage d'une fonction en parametre: En PYTHON, une variable peut referencer une fonction, donc, on peut transmettre une fonction comme parametre.
 
     - Exemple:

```python
from math import sin, pi

def f(x):
    return 2*x + 1

def g(x):
    return sin(x)

def h(fonc, x):
    return fonc(x)

print(f(2), h(f, 2))
print(g(pi/2), h(g, pi/2))
```

- Exemple: Fonction valeur absolue
La fonction valeur absolue est predefinie en PYTHON par $abs()$. Cet exemple illustre comment est-elle implementee:

```python
x = -7
print(abs(x))
```

La fonction valeur absolue est definie de la facon suivante:
\begin{equation}
|x| = 
\left\{ \begin{matrix}
x & si\:\:x \geq 0\\ 
-x & sinon
\end{matrix} \right.
\end{equation}

```python
def Fonc_valeur_absolue(x):
    if x >= 0:
        return x
    else:
        return -x
Fonc_valeur_absolue(-8) == abs(-8)
```

## b- Modules simples:
Souvent, On collecte les fonctions dont on a besoin dans un script PYTHON, ce qui cree un module avec des fonctionnalites PYTHON supplementaires ! 
- Exemple: Un script appelle "fonctions.py", le corps de cette fonction est comme-suit:

```python
def f(x):
    return 2*x + 1
def g(x):
    return x**2 + 4*x - 5
def h(fonc, x):
    try:
        return 1/fonc(x)
    except ZeroDivisionError:
        return print("Division par Zero Impossible !")
```

```python
exec(open("fonctions.py").read())
```

On peut a present utiliser les fonctions definies dans "fonctions.py" dans un script externe ou dans un environnement IPython externe. Ce script cree porte le nom de $\textbf{Module}$.

```python
print(f(2), g(2), h(f,5))
```
